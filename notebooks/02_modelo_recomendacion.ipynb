```python
import pandas as pd
```

```python
df= pd.read_csv('../data/movies_cleaned.csv')
display(df.head())
```

# Filtrado colaborativo usando la correlacion de pearson

```python
user_movie_matrix = df.pivot_table(index='userId', columns='title', values='rating')
```

```python
# Suponiendo que ya tienes tu 'user_movie_matrix'
star_wars_ratings = user_movie_matrix['Star Wars (1977)']

# Método Pearson (El clásico)
sim_pearson = user_movie_matrix.corrwith(star_wars_ratings, method='pearson')

# Método Spearman (Tu propuesta estadística)
sim_spearman = user_movie_matrix.corrwith(star_wars_ratings, method='spearman')

# Comparar resultados
df_comp = pd.DataFrame({
    'Pearson': sim_pearson,
    'Spearman': sim_spearman
}).dropna()

display(df_comp.sort_values('Spearman', ascending=False).head(10))
```
